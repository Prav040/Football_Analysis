```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import requests
from bs4 import BeautifulSoup
import json
import os
from glob import *
```

```python
path = os.getcwd()
path = path +'\\xG_Bundesliga\\Raw_Player_Stats\\'

dir = path
dateiname = []

for file in glob(dir + "/*.xlsx"):
   dateiname.append(file.split('.xlsx')[0].rpartition('_')[-1])
namen = np.array(dateiname, dtype=int)
int_low = np.max(namen)


dfs = []
#for i in range(int_low+1,int_low+9,1):
for i in range(17516,17822,1):

    id=i
    base_url = 'https://understat.com/match/'
    match = str(id)
    url = base_url + match
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'lxml')
    scripts = soup.find_all('div', {"class" : "progress-value"})

    dfs.append(stats(scripts))




new_df = pd.concat(dfs)


path = os.getcwd()
path = path +'\\xG_Bundesliga\\Raw_Match_Stats\\'

new_df.to_csv(path + '2021_match_stats.csv')
new_df.to_excel(path + '2021_match_stats.xlsx')
```

```python
def stats(scripts):
    teams = []
    goals = []
    xG = []
    shots = []
    shots_target = []
    xPTS = []
    match_id = [match, match]
    match_result = []
    act_points = []

    for i in range(0,19):
        if i == 0:
            teams.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 1: 
            teams.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 5:
            goals.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 6:
            goals.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 7:
            test = str(scripts[i]).split('progress-value">')[1].split('<span')[0] + str(scripts[i]).split('"progress-value-decimal">')[1].split('</span>')[0]
            xG.append(test)
        if i == 8:
            test = str(scripts[i]).split('progress-value">')[1].split('<span')[0] + str(scripts[i]).split('"progress-value-decimal">')[1].split('</span>')[0]
            xG.append(test)
        if i == 9:
            shots.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 10:
            shots.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 11:
            shots_target.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 12:
            shots_target.append(str(scripts[i]).split('progress-value">')[1].split('</div')[0])
        if i == 17:
            test = str(scripts[i]).split('progress-value">')[1].split('<span')[0] + str(scripts[i]).split('"progress-value-decimal">')[1].split('</span>')[0]
            xPTS.append(test)
        if i == 18:
            test = str(scripts[i]).split('progress-value">')[1].split('<span')[0] + str(scripts[i]).split('"progress-value-decimal">')[1].split('</span>')[0]
            xPTS.append(test)

    if int(goals[0]) > int(goals[1]):
        match_result.append('W') 
        match_result.append('L')
        act_points.append(3)
        act_points.append(0)
    if int(goals[0]) < int(goals[1]):
        match_result.append('L') 
        match_result.append('W')
        act_points.append(0)
        act_points.append(3)
    if int(goals[0]) == int(goals[1]):
        match_result.append('D') 
        match_result.append('D')
        act_points.append(1)
        act_points.append(1)
    col_name = ['team', 'match_result', 'goals', 'xG', 'shots', 'shots_on_Target', 'actual_Points', 'xPTS', 'match_id']
    df = pd.DataFrame([teams, match_result,goals,xG,shots,shots_target, act_points, xPTS, match_id], index=col_name)
       
    return df.T
```
