```python
import pandas as pd
from mplsoccer.pitch import Pitch
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = os.getcwd()
df = pd.read_csv(path + '\\xG_Bundesliga\\Fact_Players\\Bundesliga_xG.csv', index_col=0)
```

```python
df
```

```python
pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
df['x'] = df['x'].apply(lambda x: x*120)
df['y'] = df['y'].apply(lambda x: x*80)
```

```python

df_neu = df[(df['player'] == 'Erling Haaland') & (df['result'] == 'Goal')] 
df_neu
```
